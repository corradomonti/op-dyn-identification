```python
eps = 0.25
mu = 0.1
beta = 15
# T = 10000
# N = 50

murange = np.linspace(0, 0.5, 501)
generation_seed = 1

n_ic = 100
n_trials = 100
t_range = np.array([1000, 5000, 10000, 20000]) # 
n_range = np.array([100, 500, 1000])
```

```python
dfs = []
for c, T in enumerate(t_range):
    dfs_singleN = []
    for ct, N in enumerate(n_range):
        try:
            df_tmp = pd.read_csv(f'../outputs/mle_eps_over_N_T_spminimize_{n_trials}_{eps}_{mu}_{beta}_{N}_{T}_{generation_seed}.csv', index_col=0)
            df_tmp = df_tmp.stack().reset_index()
            df_tmp['N'] = N
            if N==100:
                df_tmp['T'] = T-300
            elif N==500:
                df_tmp['T'] = T
            elif N==1000:
                df_tmp['T'] = T+300
            else:
                df_tmp['T'] = T-250
            df_tmp['eps'] = eps
            df_tmp['mu'] = mu
            df_tmp.columns = ['x0i', 'trial', 'eps_hat', 'N', 'T', 'eps', 'mu']
            df_tmp['eps_err'] = df_tmp.eps_hat - df_tmp.eps
            dfs_singleN.append(df_tmp)
        except:
            continue
    if len(dfs_singleN)>0:
        dfs.append(pd.concat(dfs_singleN))
```

```python
len(dfs)
```

```python
dfs = pd.concat(dfs)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.lineplot(data=dfs, x='T', y='eps_err', hue='N', errorbar=('sd', 1), err_style='bars', marker='o', linestyle='--', palette='tab10')
plt.axhline(0., ls='--', c='r')
plt.ylim([-0.1, 0.1])
plt.ylabel('Estimation error')
plt.yticks([-0.1, -0.05, 0., 0.05, 0.1])
plt.xticks([0, 5000, 10000, 15000, 20000])
# plt.xlim([0., 1])
sns.despine()
plt.tight_layout()
plt.savefig(f'../figs/eps_over_N_T_mu_{mu}_eps_{eps}_beta_{beta}_err.pdf')
```
